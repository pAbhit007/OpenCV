```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
img = cv2.imread('/content/dhoni.jpg')
img
```

```python
img_x=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_x
```

```python
type(img)
```

```python
#converting images into grayscale
cv2.cvtColor(img_x,cv2.COLOR_RGB2GRAY)
```

```python
img_x.shape
```

```python
#resizing the image
image = cv2.resize(img_x,(256,256))
image
```

```python
#DIVIding all the channels seperately

plt.subplot(1,3,1)
imgBlue = image[:,:,0]
plt.imshow(imgBlue)
plt.title('Red Channel')

plt.subplot(1,3,2)
imgGreen = image[:,:,1]
plt.imshow(imgGreen)
plt.title('Green Channel')

plt.subplot(1,3,3)
imgRed = image[:,:,2]
plt.imshow(imgRed)
plt.title('Blue Channel')

plt.tight_layout()  # Adjusts subplot params so that subplots are nicely fit in the figure
plt.show()
```

```python
#using numpy showing image
import numpy as np

pic_arr = np.asarray(image)
plt.imshow(pic_arr)
```

```python
#flipping

plt.subplot(1,3,1)
img_resize = cv2.flip(image,0)
plt.imshow(img_resize)
plt.title('Flipped')

#cropped
plt.subplot(1,3,2)
img_cropped = image[0:75,50:125]
plt.imshow(img_cropped)
plt.title('Cropped')

#resize
plt.subplot(1,3,3)
img_resize = cv2.resize(image,(200,200))
plt.imshow(img_resize)
plt.title('Resized')

plt.tight_layout()
plt.show()
```

```python
#savingImage

cv2.imwrite('dhoni.png',img_cropped)
```

```python
#creating an Image from random pixels
imgi = np.random.randint(0,256,(512,512,3),dtype=np.uint8)
plt.imshow(imgi)
```

# Drawing Shapes

```python
img_r=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_r = cv2.resize(img_r,(256,256))
```

```python
#rectangle
plt.subplot(1,2,1)
img_rect = cv2.rectangle(img_r,pt1=(50,0),pt2 = (125,75),color=(0,255,255),thickness= 3)
plt.imshow(img_rect)

'''plt.subplot(1,3,2)
img_txts = cv2.putText(img_r,org=(0,0),fontScale=0.2,color=(0,255,255),thickness=1,lineType=cv2.LINE_AA,text='DHONI',fontFace =cv2.FONT_ITALIC)
plt.imshow(img_txts)'''

plt.subplot(1,2,2)
img_cir = cv2.circle(img_r,center=(80,50),radius=45,color=(0,255,255),thickness= 3)
plt.imshow(img_cir)

plt.tight_layout()
plt.show()
```
